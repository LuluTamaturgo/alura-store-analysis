### Importação dos dados



```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Cria listas com os preços de cada loja
precos_loja_1 = loja['Preço'].tolist()
precos_loja_2 = loja2['Preço'].tolist()
precos_loja_3 = loja3['Preço'].tolist()
precos_loja_4 = loja4['Preço'].tolist()
```

```python
# Calcula o faturamento de cada loja
faturamento_loja_1 = sum(precos_loja_1)
faturamento_loja_2 = sum(precos_loja_2)
faturamento_loja_3 = sum(precos_loja_3)
faturamento_loja_4 = sum(precos_loja_4)
```

```python
# Exibe os resultados
print(f'Faturamento Loja 1: R$ {faturamento_loja_1:,.2f}')
print(f'Faturamento Loja 2: R$ {faturamento_loja_2:,.2f}')
print(f'Faturamento Loja 3: R$ {faturamento_loja_3:,.2f}')
print(f'Faturamento Loja 4: R$ {faturamento_loja_4:,.2f}')
```

# 2. Vendas por Categoria


```python
# Conta a quantidade de vendas por categoria em cada loja
vendas_categoria_loja_1 = loja['Categoria do Produto'].value_counts()
vendas_categoria_loja_2 = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja_3 = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja_4 = loja4['Categoria do Produto'].value_counts()
```

```python
# Exibe os resultados
print("Vendas por Categoria - Loja 1:")
print(vendas_categoria_loja_1, '\n')

print("Vendas por Categoria - Loja 2:")
print(vendas_categoria_loja_2, '\n')

print("Vendas por Categoria - Loja 3:")
print(vendas_categoria_loja_3, '\n')

print("Vendas por Categoria - Loja 4:")
print(vendas_categoria_loja_4, '\n')
```

```python
# Cria os gráficos
plt.figure(figsize=(16, 10))

# Loja 1
plt.subplot(2, 2, 1)
vendas_categoria_loja_1.plot(kind='bar', color='skyblue')
plt.title('Loja 1 - Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)

# Loja 2
plt.subplot(2, 2, 2)
vendas_categoria_loja_2.plot(kind='bar', color='lightgreen')
plt.title('Loja 2 - Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)

# Loja 3
plt.subplot(2, 2, 3)
vendas_categoria_loja_3.plot(kind='bar', color='salmon')
plt.title('Loja 3 - Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)

# Loja 4
plt.subplot(2, 2, 4)
vendas_categoria_loja_4.plot(kind='bar', color='orchid')
plt.title('Loja 4 - Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calcula a média das avaliações de cada loja
media_avaliacao_loja_1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja_2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja_3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja_4 = loja4['Avaliação da compra'].mean()
```

```python
# Exibe os resultados
print(f'Média de Avaliação - Loja 1: {media_avaliacao_loja_1:.2f}')
print(f'Média de Avaliação - Loja 2: {media_avaliacao_loja_2:.2f}')
print(f'Média de Avaliação - Loja 3: {media_avaliacao_loja_3:.2f}')
print(f'Média de Avaliação - Loja 4: {media_avaliacao_loja_4:.2f}')
```

```python
# Calcula a média das avaliações e armazenando o resultado em um dicionário
medias = {
    'Loja 1': loja['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

# Dados para o gráfico
lojas = list(medias.keys())
avaliacoes = list(medias.values())

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.barh(lojas, avaliacoes, color='mediumslateblue')
plt.xlabel('Média de Avaliação')
plt.title('Média de Avaliação dos Clientes por Loja')
plt.xlim(0, 5)  # considerando que a nota máxima é 5
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Exibindo os valores no gráfico
for index, valor in enumerate(avaliacoes):
    plt.text(valor + 0.05, index, f'{valor:.2f}', va='center')

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Função para plotar os produtos mais e menos vendidos
def plot_produtos_vendidos(df, nome_loja):
    contagem_produtos = df['Produto'].value_counts()

    top_5 = contagem_produtos.head(5)
    bottom_5 = contagem_produtos.tail(5)

    plt.figure(figsize=(14, 5))

    # Top 5
    plt.subplot(1, 2, 1)
    top_5.plot(kind='bar', color='seagreen')
    plt.title(f'{nome_loja} - Top 5 Produtos Mais Vendidos')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)

    # Bottom 5
    plt.subplot(1, 2, 2)
    bottom_5.plot(kind='bar', color='indianred')
    plt.title(f'{nome_loja} - 5 Produtos Menos Vendidos')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()


# Gera gráficos para cada loja
plot_produtos_vendidos(loja, 'Loja 1')
plot_produtos_vendidos(loja2, 'Loja 2')
plot_produtos_vendidos(loja3, 'Loja 3')
plot_produtos_vendidos(loja4, 'Loja 4')
```

# 5. Frete Médio por Loja


```python
# Calcula o frete médio de cada loja
frete_medio_loja_1 = loja['Frete'].mean()
frete_medio_loja_2 = loja2['Frete'].mean()
frete_medio_loja_3 = loja3['Frete'].mean()
frete_medio_loja_4 = loja4['Frete'].mean()
```

```python
# Mostra os resultados
print(f'Frete Médio - Loja 1: R$ {frete_medio_loja_1:.2f}')
print(f'Frete Médio - Loja 2: R$ {frete_medio_loja_2:.2f}')
print(f'Frete Médio - Loja 3: R$ {frete_medio_loja_3:.2f}')
print(f'Frete Médio - Loja 4: R$ {frete_medio_loja_4:.2f}')
```

```python
# Dados para o gráfico
fretes = [frete_medio_loja_1, frete_medio_loja_2, frete_medio_loja_3, frete_medio_loja_4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(fretes, labels=lojas, autopct='%.1f%%', startangle=90, colors=['#66c2a5','#fc8d62','#8da0cb','#e78ac3'])
plt.title('Distribuição do Frete Médio por Loja')
plt.axis('equal')  # Deixa o gráfico redondo

plt.show()
```
